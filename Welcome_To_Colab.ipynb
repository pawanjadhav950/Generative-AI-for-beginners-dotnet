```
import os
from google.colab import userdata
from openai import OpenAI

# Retrieve your key securely from Colab's Secrets manager
try:
    os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
    # Initialize the client (it automatically picks up the environment variable)
    client = OpenAI()
    print("✅ OpenAI Client successfully initialized!")
except Exception as e:
    print("❌ Error: Make sure your 'OPENAI_API_KEY' is added to Colab Secrets and granted access.")
```

```
# Generate a text response using the gpt-4o-mini model
try:
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a witty, helpful AI programming assistant."},
            {"role": "user", "content": "Write a 1-sentence motivational cheer for a Python coder."}
        ],
        temperature=0.7
    )

    # Print the clean text output
    print(response.choices[0].message.content)

except Exception as e:
    print(f"An error occurred: {e}")
```
